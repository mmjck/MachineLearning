```python
def countFrenq(dataFrame, classe, columnTarget):
    DFClasses = dataFrame[(dataFrame["Severidade"] == classe) & (dataFrame[columnTarget] != "?")]
    DFfreq = DFClasses.groupby(columnTarget).count()
    
    print(DFfreq)
    max_classe = DFfreq["Severidade"].idxmax()

    return max_classe
 
```

```python
max_classe0 = countFrenq(dataFrame=DFmaster, classe=0, columnTarget="BI-RADS")
max_classe1 = countFrenq(dataFrame=DFmaster, classe=1, columnTarget="BI-RADS")

max_classe0, max_classe1
```

```python

def calculateMean(dataFrame, column, classe):
    DFMeanAge = dataFrame[(dataFrame[column] == '?') & (dataFrame["Severidade"] == classe)]

    meanAge = -1

    if(DFMeanAge.empty):
        return meanAge

    DFMeanAge = dataFrame[(dataFrame[column] != '?') & (dataFrame["Severidade"] == classe)]
    meanAge = int(DFMeanAge[column].astype(int).mean())

    return meanAge
```
